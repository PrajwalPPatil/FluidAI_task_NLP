```python
!pip install pymupdf
import fitz  # PyMuPDF
import spacy
import re
from collections import defaultdict

# Load NLP model
nlp = spacy.load("en_core_web_sm")

# Function to extract text from PDF
def extract_text_from_pdf(pdf_path):
    text = ""
    with fitz.open(pdf_path) as doc:
        for page in doc:
            text += page.get_text("text") + "\n"
    return text

# Function to extract key financial insights
def extract_key_insights(text):
    sections = defaultdict(str)
    keywords = {
        "growth": ["growth", "expansion", "increase", "CAGR", "prospects"],
        "financials": ["revenue", "EBITDA", "profit", "margins", "cash flow"],
        "business_changes": ["acquisition", "merger", "capex", "investment"],
        "risks": ["challenges", "headwinds", "competition", "risk", "slowdown"],
        "strategic_moves": ["strategy", "plans", "future", "vision", "roadmap"]
    }

    for line in text.split("\n"):
        for category, words in keywords.items():
            if any(re.search(fr'\b{word}\b', line, re.IGNORECASE) for word in words):
                sections[category] += line + "\n"
                break

    return sections

# Main function to process the document
pdf_path = "/content/SJS Transcript Call.pdf"
text = extract_text_from_pdf(pdf_path)
insights = extract_key_insights(text)

# Display extracted insights
for category, content in insights.items():
    print(f"\n=== {category.upper()} ===\n{content}")
```
